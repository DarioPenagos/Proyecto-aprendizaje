```python
import json
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
results=[]
raw_vals = []
for res in os.listdir("results"):
    with open(f"results/{res}") as f:
        json_dict = json.load(f)
        results.append(json_dict)
        for v in json_dict.values():
            raw_vals.append(v)

raw_vals = np.array(raw_vals)
plt.hist(raw_vals, bins=40, density=True)
plt.title("histograma distancias")

# plt.savefig("hist_distancias.svg",dpi=600,bbox_inches="tight")
```

```python
X = np.linspace(0, 40,1000)
plt.plot(X, [(raw_vals<x).mean() for x in X])
plt.xscale("log")
plt.title("fracción de simetrías identificadas")
plt.xlabel("$\\epsilon$")

# plt.savefig("frac_sim.svg",dpi=600,bbox_inches="tight")

print(f"porcentaje encontrado con epsilon=1: {(raw_vals<1).mean()}")
```

```python
sim_vals = dict()

for dist in results:
    for k in dist.keys():
        if k not in sim_vals.keys():
            sim_vals[k]=[]
        sim_vals[k].append(dist[k])


for k in sim_vals.keys():
    plt.plot(X,[(sim_vals[k]<x).mean() for x in X])
    plt.xscale("log")
    plt.title(k)
    plt.xlabel("$\\varepsilon$")

    # plt.savefig(f"graphs/{k}_frac.svg",dpi=600,bbox_inches="tight")
    plt.show()
```
